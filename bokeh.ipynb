```python
from bokeh.plotting import figure, show

p = figure()
p.circle(
    x=[1, 2, 3, 4, 5],
    y=[6, 7, 2, 4, 5],
    size=[10, 20, 30, 15, 25],  # Les tailles sont vectorisées ici
    color=["red", "blue", "green", "purple", "orange"]  # Couleurs vectorisées
)
show(p)
```

```python
p.circle(x=10, y=10, size=20, color="navy")  # Applique "navy" à line_color et fill_color
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

p = figure(title="Exemples de Markers avec Bokeh")

# Différents markers
p.circle(x=[1, 2, 3], y=[4, 5, 6], size=15, color="blue", legend_label="circle")
p.diamond(x=[1, 2, 3], y=[3, 4, 5], size=15, color="green", legend_label="diamond")
p.triangle(x=[1, 2, 3], y=[2, 3, 4], size=15, color="red", legend_label="triangle")
p.square(x=[1, 2, 3], y=[1, 2, 3], size=15, color="purple", legend_label="square")

p.legend.title = "Type de Marker"
show(p)
```
