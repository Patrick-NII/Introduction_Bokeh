```python
from bokeh.plotting import figure, show

p = figure()
p.circle(
    x=[1, 2, 3, 4, 5],
    y=[6, 7, 2, 4, 5],
    size=[10, 20, 30, 15, 25],  # Les tailles sont vectorisées ici
    color=["red", "blue", "green", "purple", "orange"]  # Couleurs vectorisées
)
show(p)
```

```python
p.circle(x=10, y=10, size=20, color="navy")  # Applique "navy" à line_color et fill_color
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

p = figure(title="Exemples de Markers avec Bokeh")

# Différents markers
p.circle(x=[1, 2, 3], y=[4, 5, 6], size=15, color="blue", legend_label="circle")
p.diamond(x=[1, 2, 3], y=[3, 4, 5], size=15, color="green", legend_label="diamond")
p.triangle(x=[1, 2, 3], y=[2, 3, 4], size=15, color="red", legend_label="triangle")
p.square(x=[1, 2, 3], y=[1, 2, 3], size=15, color="purple", legend_label="square")

p.legend.title = "Type de Marker"
show(p)
```

```python
from bokeh.plotting import figure, show

p = figure(title="Exemple de line_color et fill_color")

# Un cercle avec contour noir et remplissage bleu
p.circle(x=1, y=1, size=50, line_color="black", fill_color="blue")

# Un autre cercle avec contour rouge et remplissage vert
p.circle(x=2, y=2, size=50, line_color="red", fill_color="green")

show(p)
```

```python
week = data.loc['2010-10-01':'2010-10-08']
s= figure(x_axis_type= 'datetime', title='Glucose Range', height=350, width=800)
s.xgrid.grid_line_color=None
s.ygrid.grid_line_alpha=.5
s.xaxis.axis_label='Time'
s.yaxis.axis_label= 'Value'

s.line(week.index, week.glucose)

show(s)
```
